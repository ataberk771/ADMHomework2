```python
top_5_apps = df.groupby('app_name').size().reset_index(name='review_count').sort_values(by='review_count', ascending=False).head(5)
top_5_app_names = top_5_apps['app_name'].tolist()
top_5_df = df[df['app_name'].isin(top_5_app_names)]
```

```python
top_5_df
```

```python
def classify_purchase_status(row):
    if row['steam_purchase']:
        return 'Acquistato su Steam'
    elif row['received_for_free']:
        return 'Ricevuto Gratuitamente'
    else:
        return 'Acquisto Esterno/Altro'

top_5_df['purchase_status'] = top_5_df.apply(classify_purchase_status, axis=1)
```

```python
purchase_counts = top_5_df.groupby(['app_name', 'purchase_status']).size().reset_index(name='review_count')
total_reviews = purchase_counts.groupby('app_name')['review_count'].sum().reset_index(name='total_review_count')
purchase_counts = purchase_counts.merge(total_reviews, on='app_name')
purchase_counts['percentage'] = (purchase_counts['review_count'] / purchase_counts['total_review_count']) * 100
print(purchase_counts)
```

```python
app_names = purchase_counts['app_name'].unique()
statuses = ['Acquistato su Steam', 'Ricevuto Gratuitamente', 'Acquisto Esterno/Altro']
percentages = {status: [] for status in statuses}

for app in app_names:
    for status in statuses:
        percentage = purchase_counts[(purchase_counts['app_name'] == app) & (purchase_counts['purchase_status'] == status)]['percentage']
        percentages[status].append(percentage.values[0] if not percentage.empty else 0)
fig, ax = plt.subplots(figsize=(12, 7))
bottom = np.zeros(len(app_names))

for status in statuses:
    ax.bar(app_names, percentages[status], label=status, bottom=bottom)
    bottom += percentages[status]

ax.set_xlabel('App')
ax.set_ylabel('Percentage of reviews')
ax.set_title('Percentage of purchase for the top 5 app')
ax.legend()

plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
app_names = purchase_counts['app_name'].unique()
statuses = purchase_counts['purchase_status'].unique()
counts = {status: [] for status in statuses}

for app in app_names:
    for status in statuses:
        count = purchase_counts[(purchase_counts['app_name'] == app) & (purchase_counts['purchase_status'] == status)]['review_count']
        counts[status].append(count.values[0] if not count.empty else 0)

bar_width = 0.25
x = np.arange(len(app_names))
fig, ax = plt.subplots(figsize=(12, 7))

for i, status in enumerate(statuses):
    ax.bar(x + i * bar_width, counts[status], bar_width, label=status)

ax.set_xlabel('App')
ax.set_ylabel('Number of reviews')
ax.set_title('Absolute Values for the top 5 app')
ax.set_xticks(x + bar_width)
ax.set_xticklabels(app_names, rotation=45, ha='right')
ax.legend()

plt.tight_layout()
plt.show()
```

The two plots represent the data about the nature of the game acquired by the users who wrote the reviews. The dataset provides information, through boolean features, of the nature of the purchase. In particular, if the game has been bought or received for free by the user. After some analysis, I've seen that there is a third option if the user who wrote the review hasn't received the game either with a purchase or for free. In this case is possible that the user bought the game from other platforms. Analyzing the graphs we can see that the percentage of users that received the games for free is a limited part, often around 1%. 20% of the users usually have bought the game from other sources and the remaining ones, representing the majority, have purchased the game on Steam. From the plot with absolute value, we can see that the measures of purchase are proportional to the number of reviews that we have seen in point 2.2., with a heavy difference between PUBG and the other games.

```python
def get_time(in_time, end_time, ora_minuti_count):
    
    ora_minuti_count['ora'] = pd.to_datetime(ora_minuti_count['ora'], format='%H:%M')
    in_time = pd.to_datetime(in_time, format='%H:%M')
    end_time = pd.to_datetime(end_time, format='%H:%M')
    
    if in_time <= end_time:
        filtered_df = ora_minuti_count[(ora_minuti_count['ora'] >= in_time) & (ora_minuti_count['ora'] <= end_time)]
    else:
        filtered_df = ora_minuti_count[(ora_minuti_count['ora'] >= in_time) | (ora_minuti_count['ora'] <= end_time)]
    return filtered_df['review_per_ora'].sum()
```

```python
# I create a sub-dataframe with only the columns of interest in the study.

sub_data = df[df['author.playtime_at_review'].notna() & df['recommended'].notna()][['author.playtime_at_review', 'recommended']]
```

```python
# I reclassify the playtime expressing it in hours.

sub_data["ore_di_gioco"] = (sub_data["author.playtime_at_review"] / 60).astype(int)
sub_data[["author.playtime_at_review", "ore_di_gioco"]].head()
sub_data.shape
```

```python
# I reclassify the dichotomous variable regarding game recommendations on a scale from 1 to 10 to simplify future analyses.

sub_data["raccomando_rescale"] = sub_data['recommended'].apply(lambda x: 10 if x else 0)
sub_data[["recommended", "raccomando_rescale"]].head()
```

```python
# Analyze the relationship between the amount of time a user has spent on an application and their review score.

sub_data["ore_di_gioco"].describe().round().astype(int)
```

```python
plt.hist(sub_data[sub_data['ore_di_gioco'] <= 250]['ore_di_gioco'], bins=50)
plt.title("distribution of play time (h)")
plt.xlabel("play time (h)")
plt.ylabel("Number of players")
plt.show()
```

In order to analyze the relationship between two variables, it is necessary to first observe the distribution of both variables. In particular, from the analysis of playtime, it appears to have a distribution with high variability,  σ=398 , and a median value of  Me=31h , which is noticeably different from the mean  μ=146h  (a value heavily influenced by the presence of outliers). Thus, it can be observed that it has a distribution with a long right tail.

```python
sub_data["raccomando_rescale"].describe().round().astype(int)
```

Observing the summary statistics of the dichotomous variable associated with the ratings, it can be seen that the vast majority of users recommend the game they are reviewing.

```python
sub_data['ore_di_gioco'].corr(sub_data['raccomando_rescale'])
```

The value of correlation quantifies on a scale from -1 to 1 how much the two variables under study influence each other. A value very close to zero, as in this case, does not indicate any apparent relationship. However, since the comparison is made between two variables expressed on markedly different scales, correlation is not the best tool to explain the relationship between these two variables.

```python
y = sub_data['ore_di_gioco']
X = sub_data['raccomando_rescale']
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
print(model.summary())
```

From the regression analysis, it can be observed that taking two individuals with equal conditions, those who review the game positively tend to play an average of 6 hours less than an individual who reviews the game negatively. Although it shows a positive index, the difference between the two categories (whether to recommend the game or not) is minimal and therefore does not indicate any significant relationship.

```python
# I create a scatter plot of the relationship between hours of play and the rating by taking a subsample in order to improve the graphical representation.
plt.figure(figsize=(8, 6))
scatterplot = sns.scatterplot(data= sub_data.iloc[:100000, :], x='raccomando_rescale', y='ore_di_gioco', alpha=0.5)

# I display the regression line so that the conclusions drawn from the analysis can be observed.
intercept, slope = model.params
plt.plot(sub_data['raccomando_rescale'], intercept + slope * sub_data['raccomando_rescale'], color='red')

plt.ylim(0,300)
plt.title("Linear regression between play time and recommended")
plt.xlabel("recommended Rescale")
plt.ylabel("Play time")

plt.show()
```

```python
# Do more experienced users (who have used the application longer) tend to give higher or lower ratings? Comment on any trends you observe.

# I divide the hours of gameplay into quartiles in order to identify and group the experienced and new users.
quartili = sub_data['ore_di_gioco'].quantile([0, 0.25, 0.5, 0.75, 1]).values
sub_data['ore_gioco_classe'] = pd.cut(sub_data['ore_di_gioco'], bins=quartili, include_lowest=True)

classi_nuove = [
    "new users",
    "Learning users",
    "Intermediate users",
    "veteran users",
]

sub_data['ore_gioco_classe'] = sub_data['ore_gioco_classe'].cat.rename_categories(classi_nuove)

# I will calculate summary statistics for each user category.
sub_data.groupby('ore_gioco_classe')['raccomando_rescale'].describe()
```

By dividing the population into four sections based on the quartile of the distribution of hours played, it is possible to categorize individuals according to their experience with the reviewed game. Although this is a matter of dependence between "a review gets at least one helpful vote" and "the reviewer has submitted at least five reviews before the current review", I will make use of some concepts of probability theory and test for possible dependence in a statistical way.

- Let **A** be the event that "a review receives at least one helpful vote."
- Let **B** be the event that "the reviewer has submitted at least five reviews prior to this review."

The aim is to verify whether events **A** and **B** are independent. Two events are said to be independent if:

$$
P(A \cap B) = P(A) \times P(B)
$$


or equivalently,

$$
P(A | B) = P(A)
$$

I would like to verify if the probability of a review receiving a helpful vote is independent of whether the reviewer has submitted five or more reviews in the past. If the following probability expressions are equal, then the events are independent.

To evaluate independence I need to compute the following probabilities:

1. P(A): The probability that a review receives at least one helpful vote, computed as the share of reviews in a dataset having one or more helpful votes.
2. P(B): The probability that the reviewer submitted at least five reviews before the current review. This involves determining the number of reviews a user had previously submitted at the time each review was written and then calculating the proportion of reviews where this count is five or higher.
3. P(A | B): The probability that a review receives at least one helpful vote, given that the reviewer has submitted at least five previous reviews. This is computed by filtering the data to include only reviews for which the reviewer had five or more previous reviews and then calculating the fraction of these reviews that received one or more helpful votes.

If 
$$
P(A | B) \approx P(A)
$$
,then we may consider events **A** and **B** as independent.

If 
$$
P(A | B) \neq P(A)
$$
,we have evidence of a dependency between these two events.

```python
# Createed a copy of the relevant columns
data = df[['author.steamid', 'timestamp_created', 'votes_helpful']].copy()

data = data.sort_values(by=['author.steamid', 'timestamp_created'])

# Calculate cumulative count of reviews for each reviewer
data['cumulative_reviews'] = data.groupby('author.steamid').cumcount()

# Check if cumulative reviews before the current review are at least 5
B = data['cumulative_reviews'] >= 4

# Check if the review received at least one helpful vote
A = data['votes_helpful'] > 0

# Calculate probabilities
P_A = A.mean()
P_B = B.mean()
P_A_given_B = data[A & B].shape[0] / B.sum()

print(f"P(A): {P_A:.4%}")
print(f"P(B): {P_B:.4%}")
print(f"P(A|B): {P_A_given_B:.4%}")

# Check for independence
if abs(P_A_given_B - P_A) < 0.01:
    print("The events A and B can be considered independent.")
else:
    print("The events A and B can be considered dependent.")
```
